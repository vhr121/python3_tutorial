Neural Networks:

components:

input, output, weights and biaz

generally follow y=mx+c   where m= weights c= biaz , our main aim is to find the weights and biaz so in the future given x we can find y.


The input is fed to the activation function along with the weights and based on the input value the activation function is triggered.


Steps of working for a single neuron:

*The input(x) is fed along with some random weight (w) in the first iteration to the neurons activation function, the function outputs a value and the error is calculated. Now the error is reduced by adjusting the weights with the help of certain algorithms like gross entropy etc,,, This process is repeated in each iteration till the error is reduced to certain acceptable value.



1. Each feature is an input 

2. The weights are chosen randomly

3. Biaz are used to shift the threshold






The inputs are multiplied by the weights and then added with the biaz, The value thus obtained is fed to the activation function like sigmoid(or tan...)


* forward Propogation:

The value obtained is the output, from this output the error is found with the help of error function like cross entropy, the collective error of all the samples is the cost which is calculated with the help of cost function. 

* Back Propogation
Our main aim is to reduce this cost, this is obtained by using the function that reduce the error like gradient descent, that calculates the new value of weights accordingly.


The above process is repeated till the error is as small as possible.
